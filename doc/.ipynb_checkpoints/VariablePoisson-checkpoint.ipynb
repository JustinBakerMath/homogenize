## Standard Poisson

```python
n = m = 21
h=1/(n-1)
xx,yy = np.meshgrid(np.linspace(0,1,n),np.linspace(0,1,m))

BC = "Mixed"

# Initial Distribution
rho = np.zeros((n,m))
rho[n//2,n//4] = 1

# Final Distribution
sigma = np.zeros((n,m))
sigma[n // 2,3*n // 4] = 1

#Gradient C
Gc = np.ones(n*n)

plt.subplot(121)
plt.imshow(rho)
plt.subplot(122)
plt.imshow(sigma)
```

```python
f = sigma.flatten()-rho.flatten()
phi = cp.Variable(n*m)
P = DD2P5(n,m,bc=BC,matrix=True)
print(P.shape)
prob = cp.Problem(cp.Minimize(cp.sum_squares(P@phi+f)))
prob.solve()
```

```python
Dx = D2xCD(n,m, BC,matrix=True)
Dy = D2yCD(n,m, BC,matrix=True)
G = np.array([Dx,Dy])
Gphi = G@phi.value
plt.quiver(xx,yy,Gphi[0],Gphi[1],color='r')
```

### Reduced Grid

```python
n = m = 21
h=1/(n-1)
xx,yy = np.meshgrid(np.linspace(0,1,n),np.linspace(0,1,m))

BC = "Mixed"

# Initial Distribution
rho = np.zeros((n,m))
rho[n//2,n//4] = 1

# Final Distribution
sigma = np.zeros((n,m))
sigma[n // 2,3*n // 4] = 1

#Gradient C
Gc = np.ones(n*n)

plt.subplot(121)
plt.imshow(rho[1:-1,1:-1])
plt.subplot(122)
plt.imshow(sigma[1:-1,1:-1])
```

```python
f = (sigma[1:-1,1:-1]-rho[1:-1,1:-1]).flatten()
phi = cp.Variable((n-2)*(m-2))
P = DD2P5(n-2,m-2,bc=BC,matrix=True)
prob = cp.Problem(cp.Minimize(cp.sum_squares(P@phi+f)))
prob.solve()
```

```python
Dx = D2xCD(n-2,m-2, BC,matrix=True)
Dy = D2yCD(n-2,m-2, BC,matrix=True)
G = np.array([Dx,Dy])
Gphi = G@phi.value
plt.quiver(xx,yy,Gphi[0],Gphi[1],color='r')
```

## Variable Coefficient Elliptic Solver

```python
n = m = 21
h=1/(n-1)
xx,yy = np.meshgrid(np.linspace(0,1,n),np.linspace(0,1,m))

BC = 'Mixed'

# Initial Distribution
rho = np.zeros((n,m))
rho[n//2,n//6] = 1
# Final Distribution
sigma = np.zeros((n,m))
sigma[n // 2,5*n // 6] = 1

#lambda
lam = np.ones((n-1)*(m-1))

plt.subplot(121)
plt.imshow(rho[1:-1,1:-1])
plt.subplot(122)
plt.imshow(sigma[1:-1,1:-1])
```

```python
f = (sigma[1:-1,1:-1]-rho[1:-1,1:-1]).flatten()
phi = cp.Variable((n-2)*(m-2))
P = VP2(n-1,m-1,lam)
prob = cp.Problem(cp.Minimize(cp.sum_squares(P@phi+f)))
prob.solve()
```

```python
Dx = D2xCD(n-2,m-2, BC,matrix=True)
Dy = D2yCD(n-2,m-2, BC,matrix=True)
G = np.array([Dx,Dy])
Gphi = G@phi.value
plt.quiver(xx,yy,Gphi[0],Gphi[1], color='r')
```

### Lambda

```python
n = m = 21
h=1/(n-1)
xx,yy = np.meshgrid(np.linspace(0,1,n),np.linspace(0,1,m))

BC = 'Dirichlet'

# Initial Distribution
rho = np.zeros((n,m))
rho[n//2,n//10] = 1
# Final Distribution
sigma = np.zeros((n,m))
sigma[n // 2,9*n // 10] = 1

#lambda
e=.2
lam = np.array([e]*(m//2)*n+[1]*n+[e]*(m//2)*(n)).reshape(n,m)

#Gradient C
Gc = np.ones(n*n)

plt.subplot(131)
plt.imshow(rho)
plt.subplot(132)
plt.imshow(sigma)
plt.subplot(133)
plt.imshow(lam)

lam = lam.flatten()
lam = np.diag(np.append(lam,lam))
print(lam.shape)
```

```python
Dx = D2xFE(n,m, BC,matrix=True)
Dy = D2yFE(n,m, BC,matrix=True)
G = np.hstack([Dx,Dy]).T
print(G.shape)
Div = np.vstack([Dx.T,Dy.T]).T
print(Div.shape)
print(lam.shape)
```

```python
f = sigma.flatten()-rho.flatten()
phi = cp.Variable(n*m)
A = Div@(lam@G)
prob = cp.Problem(cp.Minimize(cp.sum_squares(A@phi+f)))
prob.solve()
```

```python
Gphi = (G@phi.value).reshape(2,n*m)
plt.quiver(xx,yy,Gphi[0],Gphi[1])
```

```python
norm = np.array([np.linalg.norm(gphi) for gphi in Gphi.T]).reshape(n,n)
plt.imshow(norm)
plt.colorbar()
```

```python
N=10
for i in range(N):
    lam = np.zeros(n*n)
    for j,nphi in enumerate(norm.flatten()):
        if abs(1-nphi)<(N-i)/N and nphi<2:
            lam[j]=1
    plt.imshow(lam.reshape(n,n))
    plt.colorbar()
    plt.show()
    lam = np.hstack((lam,lam))
    lam = np.diag(lam)
    phi = cp.Variable(n*m)
    A = Div@(lam@G)
    prob = cp.Problem(cp.Minimize(cp.sum_squares(A@phi+f)))
    prob.solve()
    Gphi = (G@phi.value).reshape(2,n*m)
    print(np.max(Gphi))
    plt.quiver(xx,yy,Gphi[0],Gphi[1])
    plt.show()
```
